#Quantidade de filmes por gênero:

```
genres_split = data['Genre'].str.split(',', expand=True).stack().reset_index(drop=True)
```



str.split(','): Divide os gêneros que estão representados em uma string separada por vírgulas. Isso resulta em várias colunas, uma para cada gênero.
stack(): Empilha as colunas criadas por split() para transformar os gêneros em uma lista de valores, um para cada filme.

reset_index(drop=True): Reseta o índice do DataFrame e remove o índice antigo para que os dados fiquem organizados corretamente.

genres_counts = genres_split.value_counts(): Conta quantas vezes cada gênero aparece, ou seja, quantos filmes pertencem a cada gênero.

#Média de avaliação por gênero:

```
avg_rating_by_genre = data.groupby('Genre')['Rating'].mean()
```

groupby('Genre'): Agrupa os filmes de acordo com o gênero.
['Rating']: Seleciona a coluna de avaliações.
mean(): Calcula a média de avaliações para cada grupo de gênero.

#Quantidade de lançamentos por ano:

```
year_counts = data['Year'].value_counts().sort_index()
```

value_counts(): Conta quantos filmes foram lançados em cada ano.
sort_index(): Organiza os resultados pelo ano (em ordem crescente).

#Filmes com as melhores avaliações:

```
top_rated_movies = data.sort_values(by='Rating', ascending=False).head(10)
```

sort_values(by='Rating', ascending=False): Organiza os filmes pela coluna de avaliações em ordem decrescente.
head(10): Exibe os 10 filmes com as melhores avaliações.

#Filmes com as piores avaliações:

```
worst_rated_movies = data.sort_values(by='Rating', ascending=True).head(10)
```

sort_values(by='Rating', ascending=True): Organiza os filmes pela coluna de avaliações em ordem crescente.
head(10): Exibe os 10 filmes com as piores avaliações.


````python
# Análise Exploratória

# Quantidade de filmes por gênero
# A função 'str.split()' separa os gêneros de cada filme que estão representados como uma string separada por vírgulas
# A função 'stack()' empilha os gêneros, criando uma nova série com os gêneros de todos os filmes
# A função 'reset_index()' reorganiza o índice e 'drop=True' remove o índice antigo
genres_split = data['Genre'].str.split(',', expand=True).stack().reset_index(drop=True)

# 'value_counts()' conta a quantidade de ocorrências de cada gênero
genres_counts = genres_split.value_counts()

# Média de avaliação por gênero
# 'groupby()' agrupa os filmes pelo gênero
# 'mean()' calcula a média de avaliação para cada grupo (gênero)
avg_rating_by_genre = data.groupby('Genre')['Rating'].mean()

# Quantidade de lançamentos por ano
# 'value_counts()' conta o número de lançamentos por ano, e 'sort_index()' organiza os resultados por ano
year_counts = data['Year'].value_counts().sort_index()

# Filmes com as melhores avaliações
# 'sort_values()' organiza os filmes pela coluna 'Rating' em ordem decrescente
# 'head(10)' pega os 10 filmes com as melhores avaliações
top_rated_movies = data.sort_values(by='Rating', ascending=False).head(10)

# Filmes com as piores avaliações
# 'sort_values()' organiza os filmes pela coluna 'Rating' em ordem crescente
# 'head(10)' pega os 10 filmes com as piores avaliações
worst_rated_movies = data.sort_values(by='Rating', ascending=True).head(10)

# Exibindo os resultados
print(genres_counts)  # Gêneros mais populares
print(avg_rating_by_genre)  # Média de avaliação por gênero
print(year_counts)  # Quantidade de lançamentos por ano
print(top_rated_movies)  # Filmes com melhores avaliações
print(worst_rated_movies)  # Filmes com piores avaliações
````

top_10_genres = genre_counts.head(10)

plt.figure(figsize=(10, 6))

O sns.barplot() cria um gráfico de barras
* x=top_10_genres.values: Passa as quantidades de filmes (valores) para o eixo X.
* y=top_10_genres.index: Passa os gêneros (índices) para o eixo Y.

* plt.ylabel('Gênero', fontsize=12): Define o rótulo do eixo Y com a descrição "Gênero", usando fonte tamanho 12.

````python
# Selecionar os 10 gêneros mais populares
# 'head(10)' seleciona os 10 gêneros mais populares a partir do resultado da contagem de gêneros
top_10_genres = genre_counts.head(10)

# Criar o gráfico de barras para os 10 gêneros mais populares
# 'plt.figure(figsize=(10, 6))' define o tamanho da figura do gráfico
# 'sns.barplot()' cria o gráfico de barras, onde:
# 'x' recebe os valores (quantidade de filmes) e
# 'y' recebe os gêneros, com a paleta 'viridis' para a coloração das barras
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_genres.values, y=top_10_genres.index, palette='viridis')

# Configurações do gráfico
# Título do gráfico
plt.title('Top 10 Gêneros Mais Populares', fontsize=16)
# Rótulo do eixo X (quantidade de filmes)
plt.xlabel('Quantidade de Filmes', fontsize=12)
# Rótulo do eixo Y (gênero)
plt.ylabel('Gênero', fontsize=12)
# Ajuste do layout para garantir que tudo se encaixe bem
plt.tight_layout()

# Exibir o gráfico
plt.show()
````

#Explicação:
##Preparar os dados:


```
launches_per_year = data['Year'].value_counts().sort_index()
```


* value_counts(): Conta quantos filmes foram lançados por ano, ou seja, calcula a quantidade de filmes lançados em cada ano.
* sort_index(): Organiza os resultados por ano, garantindo que os dados de lançamento sejam apresentados em ordem cronológica.

##Criar o gráfico de linha:

* plt.figure(figsize=(10, 6)): Define o tamanho do gráfico. A largura é 10 unidades e a altura é 6 unidades.
* sns.lineplot(): Cria o gráfico de linha:
* x=launches_per_year.index: Define os valores do eixo X como os anos de lançamento.
* y=launches_per_year.values: Define os valores do eixo Y como a quantidade de lançamentos por ano.
* marker='o': Adiciona um marcador em cada ponto da linha para destacar a quantidade de lançamentos de cada ano.

##Configurações do gráfico:

* plt.title('Evolução dos Lançamentos por Ano', fontsize=16): Define o título do gráfico com uma fonte de tamanho 16.
* plt.xlabel('Ano de Lançamento', fontsize=12): Define o rótulo do eixo X como "Ano de Lançamento", com fonte tamanho 12.
* plt.ylabel('Quantidade de Lançamentos', fontsize=12): Define o rótulo do eixo Y como "Quantidade de Lançamentos", com fonte tamanho 12.
* plt.grid(True): Ativa a grade do gráfico para facilitar a visualização dos dados. As linhas de grade ajudam a medir e comparar facilmente os valores.
* plt.tight_layout(): Ajusta automaticamente o layout do gráfico para garantir que todos os elementos do gráfico (como título e rótulos) se ajustem corretamente no espaço da figura.

###Exibir o gráfico:

plt.show(): Exibe o gráfico na célula do Jupyter ou Google Colab. Sem essa função, o gráfico não será mostrado no ambiente interativo.

````python
# Preparar os dados
# 'value_counts()' conta o número de lançamentos por ano, e 'sort_index()' organiza os resultados por ano
launches_per_year = data['Year'].value_counts().sort_index()

# Criar o gráfico de linha
# 'plt.figure(figsize=(10, 6))' define o tamanho da figura do gráfico
# 'sns.lineplot()' cria o gráfico de linha, onde:
# 'x' recebe os anos de lançamento e 'y' recebe a quantidade de lançamentos por ano
# 'marker='o'' adiciona marcadores para cada ponto de dados
plt.figure(figsize=(10, 6))
sns.lineplot(x=launches_per_year.index, y=launches_per_year.values, marker='o')

# Configurações do gráfico
# Título do gráfico
plt.title('Evolução dos Lançamentos por Ano', fontsize=16)
# Rótulo do eixo X (ano de lançamento)
plt.xlabel('Ano de Lançamento', fontsize=12)
# Rótulo do eixo Y (quantidade de lançamentos)
plt.ylabel('Quantidade de Lançamentos', fontsize=12)
# Ativa a grade do gráfico para facilitar a leitura dos valores
plt.grid(True)
# Ajuste do layout para garantir que tudo se encaixe bem
plt.tight_layout()

# Exibir o gráfico
plt.show()
````

#Criar o gráfico de dispersão:

* plt.figure(figsize=(10, 6)): Define o tamanho da figura do gráfico, com largura de 10 unidades e altura de 6 unidades.
* sns.scatterplot(): Cria o gráfico de dispersão:
* x='Revenue (Millions)': Define a coluna Revenue (Millions) como o eixo X, que representa o orçamento do filme em milhões de dólares.
* y='Rating': Define a coluna Rating como o eixo Y, que representa a avaliação média dos filmes.
* alpha=0.6: Define a transparência dos pontos no gráfico. Com alpha=0.6, os pontos terão 60% de opacidade, permitindo que pontos sobrepostos ainda sejam visíveis.
* hue='Genre': Utiliza a coluna Genre para colorir os pontos de acordo com o gênero do filme. Isso ajuda a distinguir visualmente filmes de diferentes gêneros.
* palette='viridis': Define a paleta de cores usada para colorir os pontos. A * * paleta 'viridis' é uma escolha popular por ser perceptível a pessoas com deficiências de visão de cores.

#Configurações do gráfico:

* plt.title('Relação Entre Orçamento e Avaliação', fontsize=16): Define o título do gráfico, com a fonte do título configurada para o tamanho 16.
* plt.xlabel('Orçamento (em milhões)', fontsize=12): Define o rótulo do eixo X, descrevendo que ele representa o orçamento dos filmes em milhões, com fonte tamanho 12.
* plt.ylabel('Avaliação', fontsize=12): Define o rótulo do eixo Y, explicando que ele representa a avaliação dos filmes, com fonte tamanho 12.
* plt.legend(title='Gênero', bbox_to_anchor=(1.05, 1), loc='upper left'):
* bbox_to_anchor=(1.05, 1): Coloca a legenda fora do gráfico, à direita. O parâmetro ajusta a posição da legenda.
* loc='upper left': Posiciona a legenda no canto superior esquerdo da área onde ela será exibida.
* plt.tight_layout(): Ajusta automaticamente o layout do gráfico, garantindo que o título, rótulos e legenda não se sobreponham aos dados e ao gráfico.

##Exibir o gráfico:

* plt.show(): Exibe o gráfico na célula do Jupyter ou Google Colab. Esse comando é necessário para visualizar o gráfico em ambientes interativos.

````python
# Criar o gráfico de dispersão
# 'plt.figure(figsize=(10, 6))' define o tamanho da figura do gráfico
# 'sns.scatterplot()' cria o gráfico de dispersão:
# 'x' recebe a coluna 'Revenue (Millions)', que representa o orçamento dos filmes,
# 'y' recebe a coluna 'Rating', que representa as avaliações dos filmes,
# 'alpha=0.6' define a transparência dos pontos (0 é totalmente transparente e 1 é opaco),
# 'hue' é utilizado para colorir os pontos de acordo com o gênero dos filmes, usando a paleta 'viridis'.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='Revenue (Millions)', y='Rating', alpha=0.6, hue='Genre', palette='viridis');

# Configurações do gráfico
# Título do gráfico
plt.title('Relação Entre Orçamento e Avaliação', fontsize=16)
# Rótulo do eixo X (orçamento em milhões)
plt.xlabel('Orçamento (em milhões)', fontsize=12)
# Rótulo do eixo Y (avaliação)
plt.ylabel('Avaliação', fontsize=12)
# Exibe a legenda do gráfico, que indica a correspondência entre a cor dos pontos e os gêneros dos filmes
# 'bbox_to_anchor=(1.05, 1)' posiciona a legenda à direita do gráfico
# 'loc='upper left'' coloca a legenda no canto superior esquerdo.
plt.legend(title='Gênero', bbox_to_anchor=(1.05, 1), loc='upper left')
# Ajusta o layout do gráfico para evitar sobreposição de elementos
plt.tight_layout()

# Exibir o gráfico
plt.show()
````
